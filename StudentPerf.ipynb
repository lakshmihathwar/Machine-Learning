```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
path=r"C:\Users\Ranjitha\Downloads\Student_perf.csv"
```

```python
data=pd.read_csv(path)
data
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import RobustScaler
RS=RobustScaler()
RS
```

```python
data.describe()
```

```python
data[['StudyTimeWeekly','Absences']]=RS.fit_transform(data[['StudyTimeWeekly','Absences']])
data.head()
```

```python
X=data.drop(columns='GradeClass')
X
```

```python
y=data['GradeClass']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,test_size=0.25)
```

```python
from sklearn.linear_model import LogisticRegression
lst=LogisticRegression()
lst
```

```python
lst.fit(X_train,y_train)
```

```python
y_pred=lst.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_pred,y_test)
```
