## automating all the above steps performed using pipeline() from the sklearn library
1. import all the libraries and functions for performing data processing
2. mention the steps needed to perform automation and store all the steps in the 'steps' variable
3. import pipeline() from sklearn library and create a pipeline with 'steps' variable

### step 1

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='median')
si
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
mms
```

## step 2

```python
num_steps=[('num missing treatments',si),('num_scaling',mms)]
num_steps
```

## step 3

```python
from sklearn.pipeline import Pipeline
num_pipe=Pipeline(num_steps)
num_pipe
```

## create a column transform function to automate numerical and categorical pipeline
1. create separate all numerical and categorical variables
2. create a column transform() using sklearn library
3. apply the transformer on the data

# step 1

```python
num_features=['ApplicantIncome','CoapplicantIncome','LoanAmount']
cat_features=['Property_Area']
```

# step 2

```python
from sklearn.compose import ColumnTransformer
col_trans=ColumnTransformer(
                [
                    ('numerical_pipeline',num_pipe,num_features),
                    ('categorical_pipeline',cat_pipe,cat_features)
                ]
)
col_trans
```

# step 3

```python
data_col_trans=col_trans.fit_transform(dt_mar)
```
