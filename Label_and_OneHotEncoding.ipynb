```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
url='https://raw.githubusercontent.com/anantdivanji/LearnDigital-/main/4th%20Sem/temperature_data.csv'
```

```python
data=pd.read_csv(url)
data
```

# Observations
- Color variable contains the names of different color. ex: Red,Yellow,Blue(Unordered data).
- Temperature variable contains order among the values, like very hot,hot,warm,cold(Ordered data).

### Applying label encoding on temperature

```python
data_lab=data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le
```

```python
data_lab['Temperature']=le.fit_transform(data_lab[['Temperature']])
data_lab
```

### Manual encoding of Temperature column

```python
data_man=data.copy()
data_man
```

```python
dic={
    'Cold':0,
    'Warm':1,
    'Hot':2,
    'Very Hot':3
}
dic
```

```python
data_man['Temperature']=data_man['Temperature'].replace(dic)
data_man
```

### One hot Encoding using sklearn OneHotEncoder()
1. import OneHotEncoder() from sklearn library
2. apply the OneHotEncoder() on the desired columns
3. convert step 2 output to a dataframe
4. concatenate the original data and the step 3 output
5. drop the original text columns from the data

```python
data_ohe=data.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
ohe
```

```python
dt_ohe=ohe.fit_transform(data_ohe[['Color']])
dt_ohe
```

```python
dt_ohe.toarray()
```

```python
dt_ohe=pd.DataFrame(dt_ohe.toarray(),columns=['Blue','Red','Yellow'])
dt_ohe
```

```python
data=pd.concat([data,dt_ohe],axis=1)
data
```

```python
data=data.drop(columns='Color')
data
```
